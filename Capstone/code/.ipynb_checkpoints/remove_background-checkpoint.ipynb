```python
from rembg.bg import remove
import numpy as np
import io
from PIL import Image
from PIL import ImageFile
import os
import pandas as pd
```

# 2. Remove Background

```python
# Load list of insta accounts
final_acct_df = pd.read_csv('../data/final_acct_df.csv')
```

```python
accounts = final_acct_df['handle'].tolist()
accounts
```

```python
#create empty subfolders
for subfolder_name in accounts:
    os.makedirs(os.path.join('../pictures/wo_background', subfolder_name))
```

```python
#remember to delete all DS_Store files (not visible in directory) as it will appear corrupted and code cannot run
#just go to terminal, navigate to the right folder, and type "find . -name ".DS_Store" -delete"
for account in accounts:

    def main():
        # path of the folder containing the raw images
        in_path ='../pictures/filtered_renamed_copy/' + account

        # path of the folder that will contain the modified image
        out_path ='../pictures/wo_background/'+ account

        for image_path in os.listdir(in_path):
            # image_path contains name of the image 
            input_path = os.path.join(in_path, image_path)

            # Uncomment the following line if working with truncated image formats (ex. JPEG / JPG)
            ImageFile.LOAD_TRUNCATED_IMAGES = True

            f = np.fromfile(input_path)
            try:
                result = remove(f)
                img = Image.open(io.BytesIO(result)).convert("RGBA")
            except:
                pass
        

            # :-4 from image path to remove .jpg
            #add .png to convert file to png as transparent background does not work on jpg
            full_out_path = os.path.join(out_path, image_path[:-4]+'.png')
            # full_out_path contains the path of the output
            # image that needs to be generated
            img.save(full_out_path)

            print(full_out_path)


    # Driver Function
    if __name__ == '__main__':
        main()
```
