```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import datetime

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances
from PIL import Image
from PIL import ImageFile
import os
import numpy as np
import pandas as pd



```

```python
test_accounts = ['aureliang_','speishi']
```

```python
rgba_ravel_list = []
image_list = []
instagram_handle = []
date_post = []
seq_post_samedate = []
```

```python

new_size = (400,400)
for account in test_accounts:
    in_path ='../pictures/test_wo_background/' + account
    for image_path in os.listdir(in_path):
        # image_path contains name of the image 
        input_path = os.path.join(in_path, image_path)
        imgfile = Image.open(input_path)
        imgfile = imgfile.resize(new_size)
        na = np.array(imgfile.getdata(), np.uint8)
        row = na.ravel()

        rgba_ravel_list.append(row)
        image_list.append(input_path)
        instagram_handle.append(account)
        date_post.append(image_path[-16:-6])
        seq_post_samedate.append(image_path[-5:-4])
        
    
    
```

```python
rgba_df = pd.DataFrame(zip(instagram_handle, image_list, date_post, seq_post_samedate, rgba_ravel_list), \
                       columns = ['instagram_handle', 'image_list', 'date_post', 'seq_post_samedate', 'rgba_ravel_list'])

rgba_df['date_post'] = pd.to_datetime(rgba_df['date_post'])
rgba_df['seq_post_samedate'] = rgba_df['seq_post_samedate'].astype(str).astype(int)
rgba_df
```

```python
def cluster2(x):
    x = x.reshape(-1, 4)
    x = np.array([f for f in x if f[3] !=0], np.uint8)
    clt = KMeans(n_clusters = 2, max_iter=500, random_state=5).fit(x)
    centers = clt.cluster_centers_
    centers = np.uint8(centers)
    centers_ravel = centers.ravel()
    return centers_ravel
```

```python
def cluster2_score(x):
    x = x.reshape(-1, 4)
    x = np.array([f for f in x if f[3] !=0], np.uint8)
    clt = KMeans(n_clusters = 2, max_iter=500, random_state=5).fit(x)
    score = silhouette_score(x, clt.labels_, metric='euclidean')
    return score
```

```python
def palette(centers):
    height = 50
    width = 300
    palette = np.zeros((height, width, 4), np.uint8)
    rgba_centers = centers.reshape(-1,4)
    steps = width/rgba_centers.shape[0]
    for idx, centers in enumerate(rgba_centers): 
        palette[:, int(idx*steps):(int((idx+1)*steps)), :] = centers
    return palette
```

```python
def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(10,10))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()
```

```python
def cluster2_combined(x_col):
    #get number of rows
    row_count = len(x_col)
    end_row = row_count + 1
    x = np.concatenate(x_col.iloc [0:end_row])
    x = x.reshape(-1, 4)
    x = np.array([f for f in x if f[3] !=0], np.uint8)
    clt = KMeans(n_clusters = 2, max_iter=500, random_state=5).fit(x)
    centers = clt.cluster_centers_
    centers = np.uint8(centers)
    centers_ravel = centers.ravel()
    return centers_ravel
```
