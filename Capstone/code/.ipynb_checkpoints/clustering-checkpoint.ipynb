## 2. Open image and obtain relevant RGBA code

```python
# Open image
imgfile = Image.open("../pictures/removed_background/noBG_lovebonito_test.png")

# Only pass through non-transparent pixels, i.e. those where A!=0 in the RGBA quad
na = np.array([f for f in imgfile.getdata() if f[3] !=0], np.uint8)
```

```python
na
```

## 3. K-means clustering

```python
clt = KMeans(n_clusters = 2, max_iter=500, random_state=5)
clt.fit(na)
```

```python
x = clt.cluster_centers_
x
```

```python
def palette(clusters):
    height = 50
    width = 300
    palette = np.zeros((height, width, 4), np.uint8)
    steps = width/clusters.cluster_centers_.shape[0]
    for idx, centers in enumerate(clusters.cluster_centers_): 
        palette[:, int(idx*steps):(int((idx+1)*steps)), :] = centers
    return palette
```

```python
def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(10,10))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()
```

```python
show_img_compar(imgfile, palette(clt))
```
