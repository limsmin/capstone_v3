```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
```

```python
# final list of photos
final_photos = pd.read_csv('../data/cluster_by_photo.csv')
final_photos.head()
```

```python
# list of accounts and followers
final_accounts = pd.read_csv('../data/final_acct_df.csv')
final_accounts.head()
```

```python
#plot number of photos per instagram handle
plt.figure(figsize=(9,12))
photo_account = final_photos['instagram_handle'].value_counts(ascending = True)
photo_account.plot(kind='barh')
plt.tick_params(labelsize= 15)
plt.title('Number of photos per user', fontsize = 17)

print("Average number of photos per handle: ")
print(round(statistics.mean(final_photos['instagram_handle'].value_counts()),1))
print("Median number of photos per handle: ")
print(round(statistics.median(final_photos['instagram_handle'].value_counts()),1))
```

```python
#save value_counts of photos per user into dataframe
photo_account = pd.DataFrame(photo_account)
photo_account.reset_index(inplace = True)
photo_account.rename(columns={'index': 'handle', 'instagram_handle': 'photo_count'}, inplace = True)
photo_account
```

```python
#combine number of photos with follower count
photo_account_follower = photo_account.merge(final_accounts, how = 'left', on = 'handle')
photo_account_follower
```

```python
#scatterplot of number of photos vs follower size
plt.figure(figsize=(9,9))
plt.scatter(x=photo_account_follower['follower_count'], y=photo_account_follower['photo_count'])
plt.tick_params(labelsize= 15)
plt.title('Number of photos vs follower size', fontsize = 17)
plt.xlabel('follower size in millions', fontsize = 15)
plt.ylabel('number of photos posted', fontsize = 15);
```

```python
#dataframe excluding ec24m, who is an outlier
photo_account_follower_excl_ec24m = photo_account_follower[:-1]
photo_account_follower_excl_ec24m
```

```python
#scatterplot of number of photos vs follower size, excluding outlier ec24m
plt.figure(figsize=(9,9))
plt.scatter(x=photo_account_follower_excl_ec24m['follower_count'], y=photo_account_follower_excl_ec24m['photo_count'])
plt.tick_params(labelsize= 15)
plt.title('Number of photos vs follower size', fontsize = 17)
plt.xlabel('follower size', fontsize = 15)
plt.ylabel('number of photos posted', fontsize = 15);
```

```python
#dataframe of photo count per month
photo_month = final_photos['month'].value_counts(ascending = True)
photo_month = pd.DataFrame(photo_month)
photo_month.reset_index(inplace = True)
photo_month.rename(columns={'index': 'month', 'month': 'photo_count'}, inplace = True)
photo_month = photo_month.sort_values(by=['month'], ascending = False)
#exclude december as it is an incomplete month
photo_month = photo_month.loc[photo_month['month'] != 12]
photo_month
```

```python
#barchart for number of photos per month
plt.figure(figsize=(9,9))
plt.bar(photo_month['month'], photo_month['photo_count'])
plt.xticks(np.arange(min(photo_month['month']), max(photo_month['month'])+1, 1))
plt.tick_params(labelsize= 15)
plt.title('Number of photos per month', fontsize = 17)
plt.xlabel('month from Jan-Nov', fontsize = 15)
plt.ylabel('number of photos posted', fontsize = 15);
```
