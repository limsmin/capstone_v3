```python
import numpy as np
import pandas as pd
import cv2

import PIL
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances
```

```python
cluster_df=pd.read_csv('../data/cluster_colorname.csv')
cluster_df.columns
```

```python
def converter(instr):
    return np.fromstring(instr[1:-1],dtype=int,sep=' ')
```

```python
col_list = ['image_list', 'instagram_handle', 'date_post',\
             'seq_post_samedate', 'month', 'center1_2', 'center1_3', 'center1_4', 'center2_2',\
             'center2_3', 'center2_4', 'center3_3', 'center3_4', 'center4_4', 'colorgroup1_2',\
            'colorgroup2_2','colorgroup1_3','colorgroup2_3','colorgroup3_3','colorgroup1_4',\
            'colorgroup2_4','colorgroup3_4','colorgroup4_4']

cluster_df=pd.read_csv('../data/cluster_colorname.csv',\
                       converters={'center1_2':converter, \
                                   'center1_3':converter, \
                                   'center1_4':converter,\
                                   'center2_2':converter,\
                                   'center2_3':converter,\
                                   'center2_4':converter,\
                                   'center3_3':converter,\
                                   'center3_4':converter,\
                                   'center4_4':converter,\
                                  }, usecols = col_list)

cluster_df.columns
```

```python
cluster_df.head()
```

```python
#test out on february, 3 clusters per photo
feb_df = cluster_df.loc[cluster_df['month'] == 2]
feb_df = feb_df[['image_list', 'instagram_handle', 'date_post',\
                 'seq_post_samedate', 'month', 'center1_3','center2_3','center3_3',\
                 'colorgroup1_3','colorgroup2_3','colorgroup3_3']]
feb_df.reset_index(drop=True, inplace = True)
feb_df
```

```python
#get feb consolidated array
#instantiate an array and run for loop to concatenate
feb_array = feb_df['center1_3'][0]
for n in range (1, len(feb_df)):
        feb_array = np.concatenate([feb_array, feb_df['center1_3'][n]])
for n in range (0, len(feb_df)):
        feb_array = np.concatenate([feb_array, feb_df['center2_3'][n]])
for n in range (0, len(feb_df)):
        feb_array = np.concatenate([feb_array, feb_df['center3_3'][n]])
        
print(len(feb_array))
feb_array
```

```python
feb_array = feb_array.reshape(-1,3)
feb_array
len(feb_array)
```

```python
#get list of color groups starting from 1_3
feb_colorgroup_cluster3 = [*feb_df['colorgroup1_3'], *feb_df['colorgroup2_3'], *feb_df['colorgroup3_3']] 
len(feb_colorgroup_cluster3)
```

```python
feb_array_colorgroup = pd.DataFrame(zip(feb_array, feb_colorgroup_cluster3), \
                       columns = ['rgb', 'color_group'])
feb_array_colorgroup
```

```python
feb_array_colorgroup['color_group'].value_counts()
```

```python
colorgroup_exclude = ['brown','black','nude','grey']

feb_array_colorgroup_include = feb_array_colorgroup[~feb_array_colorgroup.color_group.isin(colorgroup_exclude)]
feb_array_colorgroup_include.reset_index(drop=True, inplace = True)
feb_array_colorgroup_include
```

```python
filtered_feb_array = feb_array_colorgroup_include['rgb'][0]
for n in range (1, len(feb_array_colorgroup_include)):
        filtered_feb_array = np.concatenate([filtered_feb_array, feb_array_colorgroup_include['rgb'][n]])
```

```python
filtered_feb_array = filtered_feb_array.reshape(-1,3)
filtered_feb_array.shape
```

```python
clt = KMeans(n_clusters = 5, max_iter=500, random_state=5).fit(filtered_feb_array)
centers = clt.cluster_centers_
centers = np.uint8(centers)
centers
```

```python
def palette(centers):
    height = 50
    width = 300
    palette = np.zeros((height, width, 3), np.uint8)
    steps = width/centers.shape[0]
    for idx, centers in enumerate(centers): 
        palette[:, int(idx*steps):(int((idx+1)*steps)), :] = centers
    return palette
```

```python
def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(10,10))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()
```

```python
show_img_compar(palette(centers), palette(centers))
```

```python
score = silhouette_score(filtered_feb_array, clt.labels_, metric='euclidean')
score
```
