```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import PIL
%matplotlib inline
from sklearn.cluster import KMeans
from PIL import Image
```

```python
def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(10,10))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()
```

```python
img = cv.imread("../pictures/removed_background/noBG_lovebonito_test.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_2 = cv.imread("../pictures/removed_background/noBG_insta_test.png")
img_2 = cv.cvtColor(img_2, cv.COLOR_BGR2RGB)


# dim = (500, 300)
# # resize image
# img = cv.resize(img, dim, interpolation = cv.INTER_AREA)
# img_2 = cv.resize(img_2, dim, interpolation = cv.INTER_AREA)

show_img_compar(img, img_2)
```

```python
clt = KMeans(n_clusters = 5, max_iter=500, random_state=5)
#reshape array with 3 columns: for R, G and B
#-1 indicates unknown number of rows in order to be compatible with original shape
clt.fit(img.reshape(-1, 3))
```

```python
x = clt.cluster_centers_
x
```

```python
x.astype('uint8')
```

```python
#define a palette function
#np.uint8 is to convert into an image datatype with range 0 to 255, to avoid distorting image intensities
#create an image with a height of 50, and a width of 300 pixels to display the color groups/palette
#steps is the width of the palette image that each cluster center will take up
#shape indicates no. of rows, which is no. of clusters
def palette(clusters):
    height = 50
    width = 300
    palette = np.zeros((height, width, 3), np.uint8)
    steps = width/clusters.cluster_centers_.shape[0]
    for idx, centers in enumerate(clusters.cluster_centers_): 
        palette[:, int(idx*steps):(int((idx+1)*steps)), :] = centers
    return palette
```

```python
clt_1 = clt.fit(img.reshape(-1, 3))
show_img_compar(img, palette(clt_1))

clt_2 = clt.fit(img_2.reshape(-1, 3))
show_img_compar(img_2, palette(clt_2))
```
