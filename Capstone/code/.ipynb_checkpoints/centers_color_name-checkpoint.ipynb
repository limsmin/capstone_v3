```python
import numpy as np
import pandas as pd
import cv2

import PIL
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
# Open image
imgfile = Image.open("../pictures/wo_background/aureliang_/aureliang__2021-01-15_2.png")

# Only pass through non-transparent pixels, i.e. those where A!=0 in the RGBA quad
na = np.array([f for f in imgfile.getdata() if f[3] >200], np.uint8)
```

```python
df = pd.read_csv('../data/colors140.csv')
df.head()
```

```python
def get_color_name(R, G, B):
    minimum = 1000
    for i in range(len(df)):
        d = abs(R - int(df.loc[i, 'R']))+abs(G - int(df.loc[i, 'G']))+abs(B - int(df.loc[i, 'B']))
        if d <= minimum:
            minimum = d
            cname = df.loc[i,'color_name']
    return cname
```

```python
def palette(centers):
    height = 50
    width = 300
    palette = np.zeros((height, width, 4), np.uint8)
    steps = width/centers.shape[0]
    for idx, centers in enumerate(centers): 
        palette[:, int(idx*steps):(int((idx+1)*steps)), :] = centers
    return palette
```

```python
def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(10,10))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()
```

```python
clt = KMeans(n_clusters = 3, max_iter=500, random_state=5)
clt.fit(na)
```

```python
centers = clt.cluster_centers_
centers = np.uint8(centers)
centers
```

```python
palette(centers)
```

```python
show_img_compar(imgfile, palette(centers))
```
